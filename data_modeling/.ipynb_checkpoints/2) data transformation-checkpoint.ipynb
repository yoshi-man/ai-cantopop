# Data Transformation - Getting the data ready for training

After cleaning away unwanted characters, and formatting them into the right format, we ultimately want a txt file that we can use to feed into our model for training.

In the process, I've also taken a bit of time to look into some features of the lyrics of cantopop.

```python
numLyrics = int(cur.execute('SELECT COUNT(*) FROM lyrics').fetchall()[0][0])
print('Total lyrics we have: ', numLyrics)
lines_count = []
character_count = []

for offset in range(0, (numLyrics//500 + 1)*500, 500):
    results = list(cur.execute("SELECT lyrics FROM lyrics LIMIT 500 OFFSET {}".format(offset)).fetchall())
    results = list(map(lambda x: x[0],results))
    
    for i, lyrics in enumerate(results):
        index = offset + i
        cleaned_lyrics = clean_lyrics(lyrics)
        lines_count.append(0)
        character_count.append(0)
        for line in cleaned_lyrics:
#             print('{} has {} characters'.format(line, len(line)))
            lines_count[index] += 1
            character_count[index] += len(line)    
```
